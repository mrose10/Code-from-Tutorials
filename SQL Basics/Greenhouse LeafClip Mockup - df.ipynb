```python
from plotly.offline import init_notebook_mode, iplot
import plotly.graph_objects as go
import cufflinks as cf
import pandas as pd
import numpy as np
init_notebook_mode(connected=True)

# load dataset
df = pd.read_csv(r"C:\Users\molly\Documents\Python Scripts\LeafClip Schema Fake 2days.csv")

df.head()
```

```python
df.groupby('PlantID'=1).mean()


```

```python
# Create data 
Loc_green = df.loc[df['Location'] == 'Greenhouse']['Timestamp','Fluorescence - Red']
Loc_field = df.loc[df['Location'] == 'Field']#['Fluorescence - Red']
```

```python
# Load it into a plotly format
data = [
    go.Scatter(x=Loc_green.Timestamp, y=Loc_green['Fluorescence - Red'],name = Loc_green.PlantID),
    go.Scatter(x=Loc_field.Timestamp, y=Loc_field['Fluorescence - Red'], name = 'Field'),
    
]
```

```python
list_updatemenus = [
    {'label':'Greenhouse',
     'method': 'update',
     'args': [{'visible': [True, False]}]},
    {'label': 'Field',
     'method': 'update',
     'args': [{'visible': [False,True]}]},
] 

layout = go.Layout(title = 'Mock Data',
                  updatemenus = list([dict(buttons = list_updatemenus)]))
```

```python
fig = go.Figure(data,layout)
iplot(fig)
```
