```python
import sqlite3
import pandas as pd
import numpy as np

import plotly.graph_objects as go


import plotly.express as px
```

```python
sqlite_file = r'C:\Users\molly\Documents\Python Scripts\test.db' #Point to the db

conn = sqlite3.connect(sqlite_file)

cursor = conn.cursor() #Cursor objects are the interface to the database
```

```python
query_ghs = "SELECT Timestamp, `Fluorescence-Red`, PlantID FROM Fake2days WHERE Location = 'Greenhouse'"
query_field = "SELECT Timestamp, `Fluorescence-Red`, PlantID FROM Fake2days WHERE Location = 'Field'"
```

```python
table_ghs = pd.read_sql_query(query_ghs,conn)
table_field = pd.read_sql_query(query_field,conn)
```

```python
conn.commit() #save change to database other than sending queries
conn.close()
```

```python
# Load it into a plotly format
#data = [
#    px.scatter(table_ghs, x='Timestamp', y='Fluorescence-Red',color="PlantID"),
#    px.scatter(table_field, x='Timestamp', y='Fluorescence-Red',color="PlantID")
    
#]

# Load it into a plotly format
data = [
    go.Scatter(x=table_ghs.Timestamp, y=table_ghs['Fluorescence-Red'],mode='markers',marker_size=3),
    go.Scatter(x=table_field.Timestamp, y=table_field['Fluorescence-Red'], mode='markers',marker_size=3),
    
]
```

```python
list_updatemenus = [
    {'label':'Greenhouse',
     'method': 'update',
     'args': [{'visible': [True, False]}]},
    {'label': 'Field',
     'method': 'update',
     'args': [{'visible': [False,True]}]},
] 

layout = go.Layout(title = 'Mock Data',
                  updatemenus = list([dict(buttons = list_updatemenus)]))
```

```python
fig = go.Figure(data,layout)
fig.show()
```
